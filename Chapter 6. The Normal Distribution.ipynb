**Parametric statistics assumes** that sample data comes from a population that follows a probability distribution based on a fixed set of parameters, often the Gaussian distribution (i.e. **normal distribution**).

If a data sample is not Gaussian, then the assumptions of parametric statistical tests are violated and nonparametric statistical methods must be used.

```python
# Imports
from scipy import stats
import matplotlib.pyplot as plt
import pylab
import numpy as np
import pandas as pd
import math
```

```python
# Generic functions

def display_probability_density_chart(
    x_fill_dict=None
):
    r'''
    Display probability density chart.
    '''  
    fig, ax = plt.subplots(1, 1)
        
    x = np.linspace(
        stats.norm.ppf(0.001),
        stats.norm.ppf(0.999), 
        100
    )
    ax.plot(
        x, 
        stats.norm.pdf(x),
        'r-', 
        lw=3, 
        alpha=0.6, 
        label='Normal PDF' # Probability Density Function
    )
    
    # get space
    if not x_fill_dict:
        x_fill_1 = np.linspace(-1, 1, 100)
        x_fill_2 = np.linspace(-2, 2, 100)
        x_fill_3 = np.linspace(-3, 3, 100)    

        x_fill_list = [x_fill_1, x_fill_2, x_fill_3]
        color_list = ['k', 'grey', '#cccccc']
        label_list = ['1 STD', '2 STD', '3 STD']
        alpha = '0.3'
    else:
        x_fill_list = [np.linspace(x_fill_dict.get('left'), x_fill_dict.get('right'), 100)]
        color_list = [x_fill_dict.get('color')]
        label_list = ['z-score=' + x_fill_dict.get('label')]
        alpha = '0.7'
    
    for x_fill, color, label in zip(x_fill_list, color_list, label_list):
        # color area in chart
        ax.fill_between(
            x_fill, 
            stats.norm.pdf(x_fill), 
            color=color, 
            alpha=0.3, 
            label=label
        )

    ax.legend()
```

```python
x_fill_dict = {
    'left': -3,
    'right': 1.5,
    'color': '#cccccc',
    'label': '1.5'
}
display_probability_density_chart(x_fill_dict)
```

# Application of Normal Distribution

## Normal Distribution
Is is a **continuous distribution**, not discrete.

## Standard Normal Distribution
Properties:
- μ = 0
- σ = 1

**All normally distributed variables can be transformed into standard normally distributed variables.**

    Standard score = z = (X - μ)/σ

**Standardization** is the process of putting different variables on the same scale.

Useful for solving practical problems with parameters in absolute numbers (e.g. dollars spent on buying clothes).

```python
def get_zscore_from_x(
    x,
    mean,
    std,
):
    r'''
    Standardize x value.
    '''
    return round((x-mean)/std, 2)


def get_zscore_from_probability(
    p_left=None, # probability left side of target range (if any)
    p_right=None, # probability right side of target range (if any)    
):
    r'''
    Calculate z-score from table (with ppf).
    '''
    z = 0
    if p_left and not p_right:
        z = stats.norm.ppf(p_left)
        
    elif p_right and not p_left:
        z = 1 - stats.norm.ppf(p_right)
    
    elif p_left and p_right:
        z_left = stats.norm.ppf(p_left)
        z_right = stats.norm.ppf(p_right)
        z = z_right - z_left
    
    return z


def get_probability_from_zscore(
    z_left=None, # z-score left side of target range (if any)
    z_right=None, # z-score right side of target range (if any)
    show_chart=False,
):
    r'''
    Calculate probability (with cdf).
    '''
    p = 0
    if z_left and not z_right:
        p = stats.norm.cdf(z_left)
        z_left_chart = -3
        z_right_chart = z_left
        
    elif z_right and not z_left:
        p = 1 - stats.norm.cdf(z_right)
        z_left_chart = z_right
        z_right_chart = 3
    
    elif z_left and z_right:
        p_left = stats.norm.cdf(z_left)
        p_right = stats.norm.cdf(z_right)
        p = p_right - p_left
        z_left_chart = z_left
        z_right_chart = z_right
        
    # show chart
    if show_chart:
        x_fill_dict = {
            'left': z_left_chart,
            'right': z_right_chart,
            'color': '#cccccc',
            'label': str(z_left_chart) + ' to ' + str(z_right_chart)
        }
        display_probability_density_chart(x_fill_dict)
    
    return p
```

## Mathematical methods

### Compare 1 STD, 2 STD, 3 STD with emperical rule

```python
# degrees of freedom is the number of values in the final calculation that are free to vary.
std = np.std(starbucks_branches_per_city_np, ddof=1)
mean = np.mean(starbucks_branches_per_city_np)
n = len(starbucks_branches_per_city_np)

normal_distribution_rule = [.68, .95, .997]

for std_val, normal_val in zip(range(1, 4), normal_distribution_rule):
    # data within x STD
    within_std = sum(1 for i in starbucks_branches_per_city_np if mean-std*std_val<i<mean+std*std_val)
    print(str(std_val) + ':', within_std/n, '(normal value:' + str(normal_val) + ')')
```

Rule is simple:
- Approx 68.0% of the data falls within **1** std of the mean
- Approx 95.0% of the data falls within **2** std of the mean
- Approx 99.7% of the data falls within **3** std of the mean

### Check of outliers

```python
def find_x_for_percentile(percentile, list_input):
    list_sorted = sorted(list_input)
    values_count = len(list_sorted)
    x_pos = values_count * percentile/100

    if x_pos.is_integer():
        x_pos = int(x_pos)
        x = (list_sorted[x_pos-1] + list_sorted[x_pos]) / 2
    else:
        x_pos = int(math.ceil(x_pos))
        x = list_sorted[x_pos-1]

    return int(x)


def iqr(list_input):
    # Sort list
    list_input_sorted = sorted(list_input)
    
    # Calculate quartiles
    q1_x = find_x_for_percentile(25, list_input_sorted)
    q2_x = find_x_for_percentile(50, list_input_sorted)
    q3_x = find_x_for_percentile(75, list_input_sorted)

    # Calculate IQR
    iqr = q3_x - q1_x

    # Adjust IQR according to formula (* 1.5)
    iqr_adjusted = iqr * 1.5
    reasonable_interval = [q1_x-iqr_adjusted, q3_x+iqr_adjusted]

    # Find outliers
    outlier_list = [str(x) for x in list_input_sorted if x < reasonable_interval[0] or x > reasonable_interval[1]]

    print('Q1:', q1_x)
    print('Q2:', q2_x)
    print('Q3:', q3_x)
    print('IQR:', iqr)
    print('')
    print('Reasonable interval from', reasonable_interval[0], 'to', reasonable_interval[1])
    print('Outliers:', ','.join(outlier_list))
    
    return iqr
```

## Pearson Coefficient
It is used to assess whether data is skewed.

## Formula
PC = (3(x̅-median)/s

x̅ = sample mean

s = sample STD (with ddof=1)

If not -1 < PC < 1, then data is significantly skewed.

```python
def person_coefficient(dataset):
    dataset_np = np.array(dataset)
    x = np.mean(dataset_np)
    median = np.median(dataset_np)
    s = np.std(dataset_np, ddof=1) # sample STD
    
    # Formula
    pc = 3*(x-median)/s
    
    if pc < -1 or pc > 1:
        print('Data is significantly skewed.')
    else:
        print('Data is normal.')
    
    return pc
```

# Central Limit Theorem

To find out how means of samples of a same size vary about the population mean.

## Definitions

**Sampling distribution of sample means:** distribution using means computed from all possible random samples of specific size taken from the population.

**Sampling error:** sampling measure - population measure, because sample cannot be perfect.

## Properties
1. Mean of sample means = population mean (μ)
2. STD of sample means (sXbar) < STD population (σ)
3. STD of sample means = σ/√n

## Theorem
**As sample size n increases, the shape of distribution of sample means** taken with replacement from a population of mean μ and STD σ **will approach the normal distribution**.

For calculating the z-value, the Central Limit Theorem says that the formula should consider the sampling error.

    z = (Xbar - μ)/σ/√n, with n >= 30
    
Hence, the sample size **n has an impact on the result**. This is not the case with population z-values.

# Normal Approximation to Binomial Distribution
With a binomial distribution, the larger n and the closest p to 0.5, the more the shape becomes similar to a normal distribution.

Rule: if **np and nq > 5**, then the normal distribution can be used to approximate the binomial distribution.

However, there is a **correction needed for continuity**. Reason is that continuous distribution (i.e. normal distribution) is used to approximate a discrete distribution (binomial distribution).

**The results will not be identical, therefore better use the binomial distribution for computation**. The approximation is a shortcut when work is done manually.
